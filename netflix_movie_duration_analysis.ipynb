```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
```

# Load and inspect the Netflix data

```python
# Load the file
df = pd.read_csv('netflix_data.csv')
```

```python
# Cehck the data frame
print(df.head())
```

# Subset the dataframe for Movies

```python
# Subset the dataframe such that only rows where type is a 'Movie' are preserved

# Subset the DataFrame for type "Movie"
netflix_subset = df[df["type"] == "Movie"]

# Select only the columns of interest
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]

# Check the top 5 
netflix_subset.head()
```

# Filter the subset dataframe by movie duration

```python
# Filter the netflix_movies and get only movies that areshorter than 60 minutes.

# Subset the DataFrame for type "Movie"
short_movies = netflix_subset[netflix_subset["duration"] <60]

# Select only the columns of interest
short_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]

# Check the top 5 
short_movies.head()
```

# Assign color to movie genres

```python
 # Initialize an empty list called colors to store your different color
 colors = []

#using a for loop, iterate through netflix_movies rows and append colors to the
# list based on the genres "Children", "Documentaries", "Stand-Up", and other.
for label, row in short_movies.iterrows():
    if row['genre'] == "Children":
        colors.append('yellow')
    elif row['genre'] == "Documentaries":
        colors.append('blue')
    elif row['genre'] == "Stand-up":
        colors.append('orange')
    else:
        colors.append('purple')
print(colors)
```

# Plotting the movie duration by year using the genre colors created



```python
# Creating a figure 12x8
fig = plt.figure(figsize=(12,8))


# Create a scatter plot of duration versus release_year
plt.scatter(netflix_movies.release_year, netflix_movies.duration, c=colors)

# Create a title and axis labels
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
```
